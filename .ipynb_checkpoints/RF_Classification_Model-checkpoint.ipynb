```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier

# load complete movies dataset
movies_data = pd.read_csv('movies_modeling.csv')
movies_data.head()
```

```python
movies_data.info()
```

```python
target = movies_data['success_level_ww']
target.head()
```

```python
testing_data = movies_data.drop(['domestic_gross_usd', 'worldwide_gross_usd', 'imdb_score', 'non_dom_gross_usd', 'worldwide_roi', 'domestic_roi', 'non_dom_roi', 'success_level_ww'], axis = 1)
testing_data.head()
```

```python
# either form other genre, or re-evaluate genre for movies below Horror
testing_data['genre'].value_counts()
```

```python
to_replace_rating = ["Not Rated","Unrated", "NC-17"]
testing_data[["rated"]] = testing_data[["rated"]].replace(to_replace_rating,"Other")
```

```python
to_replace_genre = ["Mystery","Family", "Sci-Fi", "Thriller", "Romance"]
testing_data[["genre"]] = testing_data[["genre"]].replace(to_replace_genre,"Other")
```

```python
data_train, data_test, target_train, target_test = train_test_split(testing_data, target, test_size=0.25, random_state=122)
```

```python
data_train.sort_index().head()
```

```python
data_train.info()
```

```python
# X_cat = movie_rating
# X_numeric = data_train
# movie_rating = data_train[['rated']].copy()
# # movie_genre = data_train[['genre']].copy()
# data_train_num = data_train.drop(['rated', 'genre'], axis=1)

# ohe = OneHotEncoder(sparse=False)
# ohe.fit(movie_rating)
# movie_rating_ohe_train = pd.DataFrame(
#     data=ohe.transform(movie_rating),
#     columns=[f"rating_{rating}" for rating in ohe.categories_[0]],
#     index=movie_rating.index
# )
# movie_rating_ohe_train.drop(['rating_Not Rated'], axis=1, inplace=True)
# movie_rating_ohe_train.head()
```

```python
data_train_num = data_train.drop(['rated', 'genre'], axis=1)
data_test_num = data_test.drop(['rated', 'genre'], axis=1)
train_rating = data_train[['rated']].copy()
train_genre = data_train[['genre']].copy()
test_rating = data_test[['rated']].copy()
test_genre = data_test[['genre']].copy()
```

```python
# print(data_train['rated'].unique())

ohe = OneHotEncoder(sparse=False)
ohe.fit(train_rating)
movie_rating_ohe_train = pd.DataFrame(
    data=ohe.transform(train_rating),
    columns=[f"rating_{rating}" for rating in ohe.categories_[0]],
    index=train_rating.index
)

movie_rating_ohe_train.drop(['rating_Other'], axis=1, inplace=True)
movie_rating_ohe_train
```

```python
data_train = pd.merge(data_train_num, movie_rating_ohe_train, left_index=True, right_index=True)
data_train.sort_index().head()
```

```python
# print(data_train['genre'].unique())
ohe = OneHotEncoder(sparse=False)
ohe.fit(train_genre)
movie_genre_ohe_train = pd.DataFrame(
    data=ohe.transform(train_genre),
    columns=[f"genre_{genre}" for genre in ohe.categories_[0]],
    index=train_genre.index
)
movie_genre_ohe_train.drop(['genre_Adventure'], axis=1, inplace=True)
movie_genre_ohe_train
```

```python
data_train = pd.merge(data_train, movie_genre_ohe_train, left_index=True, right_index=True)
data_train.sort_index().head()
```

```python
# print(data_test['rated'].unique())
ohe = OneHotEncoder(sparse=False)
test_rating = data_test[['rated']].copy()
ohe.fit(test_rating)
movie_rating_ohe_test = pd.DataFrame(
    data=ohe.transform(test_rating),
    columns=[f"rating_{rating}" for rating in ohe.categories_[0]],
    index=test_rating.index
)

movie_rating_ohe_test.drop(['rating_Other'], axis=1, inplace=True)
movie_rating_ohe_test
```

```python
data_test = pd.merge(data_test_num, movie_rating_ohe_test, left_index=True, right_index=True)
data_test.sort_index().head()
```

```python
# print(data_test['genre'].unique())
ohe = OneHotEncoder(sparse=False)
ohe.fit(test_genre)
movie_genre_ohe_test = pd.DataFrame(
    data=ohe.transform(test_genre),
    columns=[f"genre_{genre}" for genre in ohe.categories_[0]],
    index=test_genre.index
)
movie_genre_ohe_test.drop(['genre_Adventure'], axis=1, inplace=True)
movie_genre_ohe_test
```

```python
data_test = pd.merge(data_test, movie_genre_ohe_test, left_index=True, right_index=True)
data_test.sort_index().head()
```

```python
scaler = StandardScaler()

data_train_scaled = scaler.fit_transform(data_train)
data_test_scaled = scaler.transform(data_test)
```

```python
data_train_scaled = pd.DataFrame(data_train_scaled, columns=data_train.columns, index=data_train.index)
data_test_scaled = pd.DataFrame(data_test_scaled, columns=data_test.columns, index=data_test.index)
```

```python
tree_clf = DecisionTreeClassifier(criterion='gini', max_depth=5)
tree_clf.fit(data_train_scaled, target_train)
```

```python
# Feature importance
tree_clf.feature_importances_
```

```python
def plot_feature_importances(model):
    n_features = data_train.shape[1]
    plt.figure(figsize=(8,12))
    plt.barh(range(n_features), model.feature_importances_, align='center') 
    plt.yticks(np.arange(n_features), data_train.columns.values) 
    plt.xlabel('Feature importance')
    plt.ylabel('Feature')

plot_feature_importances(tree_clf)
```

```python
# Test set predictions
pred = tree_clf.predict(data_test_scaled)

# Confusion matrix and classification report
print(confusion_matrix(target_test, pred))
print(classification_report(target_test, pred))
```

```python
# 52% before scaling
print("Testing Accuracy for Decision Tree Classifier: {:.4}%".format(accuracy_score(target_test, pred) * 100))
```

```python
# Instantiate and fit a RandomForestClassifier
forest = RandomForestClassifier(n_estimators=100, max_depth=5)
forest.fit(data_train_scaled, target_train)
```

```python
# Training accuracy score
forest.score(data_train_scaled, target_train)
```

```python
# Test accuracy score
forest.score(data_test_scaled, target_test)
```

```python
plot_feature_importances(forest)
```

```python
# Instantiate and fit a RandomForestClassifier
forest_2 = RandomForestClassifier(n_estimators=5, max_features=10, max_depth=2)
forest_2.fit(data_train, target_train)
```

```python
# First tree from forest_2
rf_tree_1 = forest_2.estimators_[0]
```

```python
# Feature importance
plot_feature_importances(rf_tree_1)
```

```python
# Second tree from forest_2
rf_tree_2 = forest_2.estimators_[1]
```

```python
# Feature importance
plot_feature_importances(rf_tree_2)
```
