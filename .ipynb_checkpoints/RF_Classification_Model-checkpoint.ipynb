# Classification Model Predicting Film's Financial Success
In this notebook we will pull in the movie dataset that was created in the **Movies_Data_Modeling_Prep** notebook. This notebook will finish preparing our previously cleaned data for classification modeling using both RandomForest and XGBoost in order to properly predict the financial success of movies

DATA DICTIONARY: https://github.com/hudsop2/ai_academy_capstone?tab=readme-ov-file#data-dictionary 

To start we'll import all of our necessary libraries and load our dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import tree
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from xgboost import XGBClassifier

# load complete movies dataset
movies_data = pd.read_csv('movies_modeling.csv')
movies_data.head()
```

# Section 1: Data Prepping for Classification Modeling
In this section we will adjust some of our variables so that the model runs more efficiently. We also will drop some columns that are no longer useful in  a classification modeling context.

First let's take a look at the differences in the values between our 2 classifications. This helps us know if differences are significant enough for variables to be useful in modeling.

```python
# group fields by success classification then find the average of each group
movies_data.groupby('success_level_ww').agg('mean')
```

Here we drop a few variables that are no longer needed as they only add confusion to the model and don't have clear differences towards successful and non-successful films.

```python
# dropping fields that only add confusion to model
movies_data.drop(['month_released', 'title_length', 'runtime_minutes'], axis=1, inplace=True)
```

We still need to drop these columns as well since all this data is post-production. Our model wouldn't be very useful in predicting post-production success if it needed post-production data to do so.

```python
# dropping post-production fields to ensure predictions aren't invalid
movies_data.drop(['domestic_gross_usd', 'worldwide_gross_usd', 'imdb_score', 'non_dom_gross_usd', 
                  'worldwide_roi', 'domestic_roi', 'non_dom_roi'], axis = 1, inplace=True)
```

```python
movies_data.info()
```

When preparing our data we noticed that some columns had values with very low counts. Let's take a look at value counts for the **rated** column.

```python
# value counts for 'rated' field
movies_data['rated'].value_counts()
```

Clearly movie ratings are very heavily skewed towards R and PG-13, that's ok but the small count for NC-17, Unrated, and Not Rated could give our model issues. Let's fix this by combining them into a single value of rating **Other**.

```python
# replacing unrepresented rating categories with other value
to_replace_rating = ["Not Rated","Unrated", "NC-17"]
movies_data[["rated"]] = movies_data[["rated"]].replace(to_replace_rating,"Other")
```

Next we'll do the same but this time for genre's.

```python
# either form other genre, or re-evaluate genre for movies below Horror
movies_data['genre'].value_counts()
```

```python
# replacing unrepresented genre categories with other value
to_replace_genre = ["Fantasy", "Mystery","Family", "Sci-Fi", "Thriller", "Romance"]
movies_data[["genre"]] = movies_data[["genre"]].replace(to_replace_genre,"Other")
```

Here we separate our target and test variables. **success_level_ww** is our target and simply is a value of 1 if the movie made it's production budget back in revenue (broke even) or 0 if the movie lost money.

```python
# set target variable for model
target = movies_data['success_level_ww']
target.head()
```

```python
# set testing data for model
testing_data = movies_data.drop(['success_level_ww'], axis = 1)
testing_data.head()
```

Next we complete a train test split of the data using a testing size of 25%.

```python
# train test split of our data
data_train, data_test, target_train, target_test = train_test_split(testing_data, target, test_size=0.25, random_state=122)
```

```python
# Ensuring data is properly indexed after train_test_split
data_train.sort_index().head()
```

```python
data_train.info()
```

# Section 2: One Hot Encode Variables

Now that we have dropped unnecessary columns and prepped the **rated** and **genre** columns we need to One Hot Encode (OHE) them before using the variables for modeling. We start by pulling them out of the training and testing data.

```python
# prepping data for OHE
data_train_num = data_train.drop(['rated', 'genre'], axis=1)
data_test_num = data_test.drop(['rated', 'genre'], axis=1)
train_rating = data_train[['rated']].copy()
train_genre = data_train[['genre']].copy()
test_rating = data_test[['rated']].copy()
test_genre = data_test[['genre']].copy()
```

 First let's OHE the training data of the rating column.

```python
# OHE training data rating column

ohe = OneHotEncoder(sparse=False)
ohe.fit(train_rating)
movie_rating_ohe_train = pd.DataFrame(
    data=ohe.transform(train_rating),
    columns=[f"rating_{rating}" for rating in ohe.categories_[0]],
    index=train_rating.index
)

movie_rating_ohe_train.drop(['rating_Other'], axis=1, inplace=True)
movie_rating_ohe_train
```

Now that the training data of the rating column has been OHE we can joining it back into the training dataset.

```python
# join OHE data back into training data
data_train = pd.merge(data_train_num, movie_rating_ohe_train, left_index=True, right_index=True)
data_train.sort_index().head()
```

We continue our OHE below with the training data of the genre column.

```python
# OHE training data genre column

ohe = OneHotEncoder(sparse=False)
ohe.fit(train_genre)
movie_genre_ohe_train = pd.DataFrame(
    data=ohe.transform(train_genre),
    columns=[f"genre_{genre}" for genre in ohe.categories_[0]],
    index=train_genre.index
)
movie_genre_ohe_train.drop(['genre_Adventure'], axis=1, inplace=True)
movie_genre_ohe_train
```

And again we join the OHE data back into the training dataset.

```python
# join OHE data back into training data
data_train = pd.merge(data_train, movie_genre_ohe_train, left_index=True, right_index=True)
data_train.sort_index().head()
```

We now OHE the test data of the rating column.

```python
# OHE testing data rating column

ohe = OneHotEncoder(sparse=False)
test_rating = data_test[['rated']].copy()
ohe.fit(test_rating)
movie_rating_ohe_test = pd.DataFrame(
    data=ohe.transform(test_rating),
    columns=[f"rating_{rating}" for rating in ohe.categories_[0]],
    index=test_rating.index
)

movie_rating_ohe_test.drop(['rating_Other'], axis=1, inplace=True)
movie_rating_ohe_test
```

And again join the OHE data back into the training dataset.

```python
# join OHE data back into testing data
data_test = pd.merge(data_test_num, movie_rating_ohe_test, left_index=True, right_index=True)
data_test.sort_index().head()
```

Finally we OHE the test data of the genre column.

```python
# OHE testing data genre column

ohe = OneHotEncoder(sparse=False)
ohe.fit(test_genre)
movie_genre_ohe_test = pd.DataFrame(
    data=ohe.transform(test_genre),
    columns=[f"genre_{genre}" for genre in ohe.categories_[0]],
    index=test_genre.index
)
movie_genre_ohe_test.drop(['genre_Adventure'], axis=1, inplace=True)
movie_genre_ohe_test
```

And we finish our OHE by joining the OHE test genre data back intothe test data.

```python
# join OHE data back into testing data
data_test = pd.merge(data_test, movie_genre_ohe_test, left_index=True, right_index=True)
data_test.sort_index().head()
```

# Section 3: Building  Pipeline and GridSearch

After completing our OHE our data is now fully prepared for modeling we can build our pipeline and use GridSearch to find the optimal parameters for our model. Once this is done we will run our models for RandomForest and XGBoost and look at the accuracy of these models as well as the importance of the model features in decision making.

```python
# Construct pipeline
pipe = Pipeline([('rfc', RandomForestClassifier())])

param_range = [1, 2, 3, 4, 5]

# Set grid search params
grid_params = {
    'rfc__n_estimators': [50],  # The number of trees in the forest
    'rfc__max_depth': [15],  # The maximum depth of the tree
    'rfc__min_samples_split': [7, 8, 9, 10, 11, 12],  # The minimum number of samples required to split an internal node
    'rfc__bootstrap': [True, False]  # Whether bootstrap samples are used when building trees
}

# Construct grid search
gs = GridSearchCV(estimator=pipe,
            param_grid=grid_params,
            scoring='accuracy',
            cv=10)
```

Now we fit our classifier model on our training data using the gridsearch we just created.

```python
# fit classification model to our training data
gs.fit(data_train, target_train)
```

Once our model is fit we can then test it's ability to predict and look at the accuracy of those predictions. The best accuracy we were able to get on the testing data was 74.05% and it used the GridSearch above, with rfc_min_samples_split = 8 and rfc_bootstrap = True

```python
# predict our test data using our fitted model and check for accuracy
target_pred = gs.predict(data_test)
print(accuracy_score(target_test, target_pred))
```

The best accuracy and best parameters give us additional insight into the best accuracy score on the training data and the parameters it used to accomplish that score. These parameters can help us optimize our gridsearch.

```python
# looking at the best accuracy score and the parameters that helped us achieve that score
print('Best accuracy: %.3f' % gs.best_score_)
print('\nBest params:\n', gs.best_params_)
```

The confusion matrix we print out below gives us a better understanding of what the model is predicting outside of the accuracy. The most important data points to look at are Precision and Recall. Both are calculated for the model when trying to classify movies as 0 (failure) or 1 (success). 
- Precision = True Positives / (True Positives + False Positives)
- Recall = True Positives / (True Positives + False Negatives)

```python
# looking at data point surrounding model performance for classification
print(confusion_matrix(target_test, target_pred))
print(classification_report(target_test, target_pred))
```

This function will allow us to chart out the importance of our features within our dataset when classifying a film.

```python
# function that takes in a model and plots the importance of features used in the model's
# decision making
def plot_feature_importances(model):
    n_features = data_train.shape[1]
    plt.figure(figsize=(8,12))
    plt.barh(range(n_features), model.feature_importances_, align='center') 
    plt.yticks(np.arange(n_features), data_train.columns.values) 
    plt.xlabel('Feature importance')
    plt.ylabel('Feature')
```

Above we used GridSearch to find the best possible combination of parameters for our classifier. Now we can just run a single instance of those parameters below. 

```python
# Instantiate and fit a RandomForestClassifier
forest = RandomForestClassifier(n_estimators=50, max_depth=15, min_samples_split=8, bootstrap=True)
forest.fit(data_train, target_train)
```

Next we look at the accuracy score on the training target data.

```python
# Training accuracy score
forest.score(data_train, target_train)
```

And now we look at the accuracy score on the testing target data.

```python
# Test accuracy score
forest.score(data_test, target_test)
```

Once we have our classifier fitted we can graph the feature importance using the function we made above.

```python
# plotting the feature importance of RandomForestClassifier model
plot_feature_importances(forest)
```
