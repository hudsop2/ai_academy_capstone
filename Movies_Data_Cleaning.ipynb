```python
# load pandas library
# load date library
import pandas as pd
from datetime import date
```

```python
# load first movies dataset (https://www.kaggle.com/datasets/danielgrijalvas/movies)
movies_data = pd.read_csv('movies.csv')
movies_data.head()
```

```python
# convert year column to a string
movies_data['year'] = movies_data['year'].apply(lambda x: str(x))
```

```python
# drop duplicates of movie name and year it was released, to ensure no duplicates of movies
movies_data = movies_data.drop_duplicates(['name', 'year'])
```

```python
movies_data.info()
```

```python
# load second movies dataset (https://www.kaggle.com/datasets/alessandrolobello/the-ultimate-film-statistics-dataset-for-ml)
# Create year column based on movie production date
movies_2_data = pd.read_csv('movies_2.csv')
movies_2_data['year'] = movies_2_data['production_date'].str[:4]
movies_2_data.head()
```

```python
# drop duplicates of movie name and year it was released, to ensure no duplicates of movies
movies_2_data = movies_2_data.drop_duplicates(['movie_title', 'year'])
```

```python
movies_2_data.info()
```

```python
# load third movies dataset (https://www.kaggle.com/datasets/rounakbanik/the-movies-dataset?select=movies_metadata.csv)
# Create year column based on movie release date
movies_3_data = pd.read_csv('movies_3.csv')
movies_3_data['year'] = movies_3_data['release_date'].str[:4]
movies_3_data.head()
```

```python
# drop duplicates of movie name and year it was released, to ensure no duplicates of movies
movies_3_data = movies_3_data.drop_duplicates(['original_title', 'year'])
```

```python
movies_3_data.info()
```

```python
# movies_2_data['production_date'] = movies_2_data['production_date'].apply(lambda x: date.fromisoformat(x))
```

```python
# join first and second movies dataframes based on each movie's title and year of release, keeping only movies that exist
# in both dataframes
movies_complete = pd.merge(movies_data, movies_2_data, how='inner', 
                           left_on=['name', 'year'], 
                           right_on=['movie_title', 'year'])
movies_complete.head()
```

```python
movies_complete.info()
```

```python
# join the third movies dataframes to the previously joined first and second dataframe based on each movie's title and 
# year of release, keeping only movies that exist in both dataframes
movies_complete = pd.merge(movies_complete, movies_3_data, how='inner', 
                           left_on=['name', 'year'], 
                           right_on=['original_title', 'year'])
movies_complete.head()
```

```python
movies_complete.info()
```

```python
# filter out columns that are repeats or of no use to the analysis
movies_complete = movies_complete[['name', 'release_date', 'rating', 'genre', 'runtime_minutes', 'director_professions', 'belongs_to_collection', 'tagline', 'Production budget $', 'Domestic gross $', 'Worldwide gross $', 'score']]
movies_complete.head()
```

```python
movies_complete.info()
```

```python
# transform belongs_to_collection column to no longer track what collection a movie is a part of, but simply a boolean 
# variable listing whether or not each movie is a part of a collection or not
movies_complete['belongs_to_collection'] = (movies_complete['belongs_to_collection'].notnull()).astype('int')
movies_complete.head()
```

```python
movies_complete.info()
```

```python
# get dummies / one hot encode for: genre, rating, director_professions
```

```python
# split data into multiple columns (role1, role2, etc.)
# get_dummies for each of those columns (writer, producer, etc.)
# merge and sum columns by common name
```

```python
# split the director_professions column into 3 columns listing each role or lack of role that the director was involved in
# during the production of the movie
role1=[]
role2=[]
role3=[]
for roles in movies_complete['director_professions']:
    roles_split = roles.split(',')
    if roles_split[0] == '-':
        role1.append('')
        role2.append('')
        role3.append('')
    elif len(roles_split) == 1:
        role1.append(roles_split[0])
        role2.append('')
        role3.append('')
    elif len(roles_split) == 2:
        role1.append(roles_split[0])
        role2.append(roles_split[1])
        role3.append('')    
    elif len(roles_split) == 3:
        role1.append(roles_split[0])
        role2.append(roles_split[1])
        role3.append(roles_split[2])  
print('role1:', role1)
print('role2:', role2)
print('role3:', role3)
```

```python
director_roles = pd.DataFrame({'role1': role1, 'role2': role2, 'role3': role3})
director_roles.head()
```

```python
# transform the 3 director columns from categorical columns into label encoded columns for each role type
director_roles = pd.get_dummies(director_roles)
director_roles.head()
```

```python
director_roles.info()
```

```python
# combine the values of the columns that refer to the same categories but were split previously based on column
director_roles['dir_acted'] = director_roles[['role1_actor', 'role2_actor', 'role3_actor',
                                'role1_actress', 'role2_actress', 'role3_actress']].sum(axis=1)
director_roles['dir_animation_dept'] = director_roles[['role1_animation_department',
                                                       'role2_animation_department',
                                                       'role3_animation_department']].sum(axis=1)
director_roles['dir_art_dept'] = director_roles[['role1_art_department',
                                                 'role2_art_department',
                                                 'role3_art_department']].sum(axis=1)
director_roles['dir_assistant_dir'] = director_roles[['role1_assistant_director',
                                                      'role2_assistant_director',
                                                      'role3_assistant_director']].sum(axis=1)
director_roles['dir_camera_dept'] = director_roles[['role1_camera_department',
                                                    'role2_camera_department',
                                                    'role3_camera_department']].sum(axis=1)
director_roles['dir_casting_dir'] = director_roles[['role1_casting_director']].sum(axis=1)
director_roles['dir_cinematographer'] = director_roles[['role1_cinematographer',
                                                    'role2_cinematographer',
                                                    'role3_cinematographer']].sum(axis=1)
director_roles['dir_composer'] = director_roles[['role1_composer',
                                                    'role3_composer']].sum(axis=1)
director_roles['dir_editor'] = director_roles[['role1_editor',
                                                    'role2_editor',
                                                    'role3_editor']].sum(axis=1)
director_roles['dir_editorial_dept'] = director_roles[['role1_editorial_department',
                                                    'role2_editorial_department',
                                                    'role3_editorial_department']].sum(axis=1)
director_roles['dir_music_artist'] = director_roles[['role1_music_artist']].sum(axis=1)
director_roles['dir_music_dept'] = director_roles[['role1_music_department',
                                                    'role2_music_department',
                                                    'role3_music_department']].sum(axis=1)
director_roles['dir_producer'] = director_roles[['role1_producer',
                                                    'role2_producer',
                                                    'role3_producer']].sum(axis=1)
director_roles['dir_production_designer'] = director_roles[['role1_production_designer',
                                                    'role2_production_designer',
                                                    'role3_production_designer']].sum(axis=1)
director_roles['dir_soundtrack'] = director_roles[['role1_soundtrack',
                                                    'role2_soundtrack',
                                                    'role3_soundtrack']].sum(axis=1)
director_roles['dir_special_effects'] = director_roles[['role1_special_effects',
                                                    'role3_special_effects']].sum(axis=1)
director_roles['dir_stunts'] = director_roles[['role1_stunts']].sum(axis=1)
director_roles['dir_visual_effects'] = director_roles[['role1_visual_effects',
                                                    'role2_visual_effects',
                                                    'role3_visual_effects']].sum(axis=1)
director_roles['dir_writer'] = director_roles[['role1_writer',
                                                    'role2_writer',
                                                    'role3_writer']].sum(axis=1)
director_roles['dir_art_dir'] = director_roles[['role2_art_director',
                                                    'role3_art_director']].sum(axis=1)
director_roles['dir_casting_dept'] = director_roles[['role2_casting_department']].sum(axis=1)
director_roles['dir_costume_designer'] = director_roles[['role2_costume_designer']].sum(axis=1)
director_roles['dir_makeup_dept'] = director_roles[['role2_make_up_department']].sum(axis=1)
director_roles['dir_production_manager'] = director_roles[['role2_production_manager',
                                                    'role3_production_manager']].sum(axis=1)
director_roles['dir_script_dept'] = director_roles[['role2_script_department',
                                                    'role3_script_department']].sum(axis=1)
director_roles['dir_location_management'] = director_roles[['role3_location_management']].sum(axis=1)
director_roles['dir_sound_dept'] = director_roles[['role3_sound_department']].sum(axis=1)
```

```python
# filter out all columns besides the newly created director_roles columns
director_roles = director_roles[['dir_acted', 'dir_animation_dept', 'dir_art_dept', 'dir_art_dir', 'dir_assistant_dir', 'dir_camera_dept', 'dir_casting_dept', 'dir_casting_dir', 'dir_cinematographer', 'dir_composer', 'dir_costume_designer', 'dir_editor', 'dir_editorial_dept', 'dir_location_management', 'dir_makeup_dept', 'dir_music_artist', 'dir_music_dept', 'dir_producer', 'dir_production_designer', 'dir_production_manager', 'dir_script_dept', 'dir_sound_dept', 'dir_soundtrack', 'dir_special_effects', 'dir_stunts', 'dir_visual_effects', 'dir_writer']]
director_roles.head()
```

```python
director_roles.info()
```

```python
# combine movies_complete dataframe and director_roles dataframe
movies_complete = pd.concat([movies_complete, director_roles], axis=1)
```

```python
# drop director_professions column now that it has successfully been label encoded
movies_complete = movies_complete.drop(['director_professions'], axis=1)
movies_complete.head()
```

```python
# rename columns for more standardized and descriptive naming
movies_complete = movies_complete.rename(columns={'name': 'movie_title', 'rating': 'rated',
                                     'Production budget $': 'production_budget_usd', 
                                    'Domestic gross $': 'domestic_gross_usd',
                                    'Worldwide gross $': 'worldwide_gross_usd',
                                    'score': 'imdb_score'})
```

```python
# save the completed dataframe as a csv to be used later in analysis
movies_complete.to_csv('movies_complete.csv', index=False)
```
